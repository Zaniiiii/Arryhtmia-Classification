# Train model

```python
#mengacak sample
df = df.sample(400).reset_index(drop=True)
```

```python
#Melakukan folding pada data
fold1 = (df.iloc[0:100].reset_index(drop=True), df.iloc[150:250].reset_index(drop=True))
fold2 = (df.iloc[100:150].reset_index(drop=True), df.iloc[250:300].reset_index(drop=True))
fold3 = (df.iloc[150:200].reset_index(drop=True), df.iloc[0:100].reset_index(drop=True))


test, train = fold2
train
```

```python
# melakukan normalisasi data dengan min max norm
def norm(df):
  df = (df - df.min()) / (df.max() - df.min())
  return df

# assign fitur dan kelas
X = df.drop('diagnosis', axis = 1) # untuk fitur
Y = df.diagnosis # untuk kelas

# melakukan normalisasi untuk fitur (X)
X = norm(X)
X
```

```python
# menghitung jarak menggunakan euclidean distance
def euclidean(x1, x2):
  return np.sqrt(np.sum((x1-x2)**2))

# euclidean distance data baris pertama dan kedua
euclidean(X.iloc[0], X.iloc[1])
```

```python
# training data
def knn(X_train, y_train, X_test, k): # k = tetangga
  dist = []
  for row in range(X_train.shape[0]):
    dist.append(euclidean(X_train.iloc[row], X_test))

  data = X_train.copy()

  data['Dist'] = dist
  data['Class'] = y_train
  data = data.sort_values(by = 'Dist').reset_index(drop = True)

  # mengambil label class yang paling sering muncul
  y_pred = data.iloc[:k].Class.mode()
  return y_pred[0]
```

```python
# menghitung akurasi output berdasarkan label class
def acc(y_pred, y_true):
  true = 0
  for i in range(len(y_pred)):
    if y_pred[i] == y_true[i]:
      true += 1
  return true/len(y_pred)
```

```python
# mengevaluasi model menggunakan fold
def evaluate(fold, k):
  test, train = fold
  X_train, y_train = train.drop('diagnosis', axis = 1), train.diagnosis
  X_test, y_test = test.drop('diagnosis', axis = 1), test.diagnosis
  X_train = norm(X_train)
  X_test = norm(X_test)
  y_preds = []
  for row in range(X_test.shape[0]):
    y_preds.append(knn(X_train, y_train, X_test.iloc[row], k))
  return (acc(y_preds, y_test))
```

```python
k = 20
accs = []
folds = [fold1, fold2, fold3]
for i in range(len(folds)):
  accs.append(evaluate(folds[i], k))

print(f'Menggunakan k : {k}, dengan nilai rata-rata akurasi : {sum(accs)/3}')
```
